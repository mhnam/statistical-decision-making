## Appendix
This appendix contains additional details on the install command used to create the tprstats Python Environment.

```
# install env using a link to the tprstats-python GitHub env.yaml
!jlab env create --name tprstats --source=https://raw.githubusercontent.com/dnepple/tprstats-python/refs/heads/main/env.yaml --source-type=conda-env-file
```
This command creates an environment named tprstats. To know what packages to install in the environment, a conda-env-file(source type) needs to be included. In this instance, the conda-env-file is a weblink to the tprstats GitHub repository(source).

A copy of the conda-env-file called env.yml is included for you to look at. "channel" refers to where the Python packages for the environment should be downloaded from. We use conda-forge because it is open source. "dependencies" refers to the packages installed in the environment. This lists the packages used in the course. The last command installs the tprstats package from our GitHub.

You can use the included env.yml file to install tprstats, though it's not recommend. It involves *relative paths* which can get a bit complicated. If you were to run this command, you would need to make sure the command was run in the same directory as env.yaml.
```
# install environment using local env.yaml
!jlab env create --name tprstats --source=env.yaml --source-type=conda-env-file
```
Notice all that's changed is the source.
