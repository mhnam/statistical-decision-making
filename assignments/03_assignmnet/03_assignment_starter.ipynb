## Exercise 1: Interpreting regressions with standardized coefficients.
The objective of this exercise is to familiarize you with interpreting standardized coefficients and to provide you an interesting application of regression analysis.  

In research conducted at Carnegie Mellon, Prof. Robert Siegler played a leading role in investigating the types of early mathematics skills that build a foundation for future success in mathematics. He and his colleagues have found that the ability, by age 10, to do fractions, and the associated skills of multiplication and division, are exceedingly important in predicting children’s future success in mathematics. The table in the Word document named Siegler_Regression_Results.docx presents some of the results from their research. The columns labeled Model 1 and Model 2 investigate how well test scores in algebra at age 16 are predicted by a student’s test scores at age 10 in several mathematics skills. The effect of student IQ and demographic characteristics are also reported in the table. In the column labeled Model 1, a separate regression is run for each variable. For example, the coefficient of .42 on fractions is from a simple regression of algebra score as dependent variable and the fraction variable as the only independent variable. Model 2 is a multiple regression including all the variables in a single regression. The columns labeled Model 3 and Model 4 do the same with overall test of mathematics at age 16 as the dependent variable. Models 3 coefficients are each from a separate regression while Model 4 includes all variables at once. We will focus on Models 1 and 2, but the message of Models 3 and 4 is very similar. From the footnote, we learn that the standard error of each coefficient is in parentheses next to the coefficient. As the authors explain, the coefficients are standardized coefficients as we obtain using the scaledCoefficients() command in R. Hence, for example, the coefficient of Division in Model 2 tells us that, holding constant other variables in the model, a one standard deviation increase in the score on Division at age 10 predicts a .13 standard deviation increase in the test score in algebra at age 16.
