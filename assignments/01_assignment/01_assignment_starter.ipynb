# Mini Tutorial
Suppose we had used $Z_{1-\alpha/2} = 2.575$ in our control chart. What value of $\alpha$ corresponds to this value of $Z_{1-\alpha/2}$? We can find the probability that a random draw from the standardized normal distribution will be less than 2.575 using stats.norm.cdf(2.575). This probability corresponds to $(1-\alpha/2)$. The command stats.norm.cdf(-2.575) calculates $\alpha/2=0.005$ and hence we conclude $\alpha=0.01$.  

Suppose instead that we have chosen $\alpha$ and we want the corresponding $Z_{1-\alpha/2}$. For example, suppose we have chosen $\alpha=0.05$. Hence $(1-\alpha/2)=0.975$. We can find $Z_{1-\alpha/2}$ for $\alpha=0.05$ by substituting 0.05 in place of $\alpha$ in the command: stats.norm.ppf($1-\alpha/2$)  

**Recap**  

- `stats.norm.cdf(a)` Computes and displays $P(Z<a)$
- `1-stats.norm.cdf(b)` Computes and displays $P(Z>b)$
- `stats.norm.cdf(b) - stats.norm.cdf(a)` Computes and displays $P(a<Z<b)$
- `stats.norm.ppf(1-alpha/2)` Finds the quantile Z such that area  $\alpha/2$ lies below the lower tail and $\alpha/2$ lies above the upper tail.
